```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_csv("C:\\Data Science\\Data Science Nareshit\\EDA datasets\\visadataset.csv")
df
```

```python
sns.scatterplot(x="continent",y="yr_of_estab",data=df)
plt.title("Scatter Plot")
```

```python
#line plot
sns.lineplot(x="continent",y="yr_of_estab",data=df)
plt.title("line plot")
plt.show()
```

```python
#categorical plot
#barplot
sns.barplot(x='prevailing_wage',y='no_of_employee',data=df)
plt.title("bar plot")
plt.show()
```

```python
#box plot
sns.boxplot(x="prevailing_wage",y="case_status",data=df)
```

```python
#histogram
sns.histplot(df['continent'],bins=50,kde=True)
```

```python
#kde plot
sns.kdeplot(df['continent'],shade=True)
```
